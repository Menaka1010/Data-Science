```python
Ind_features=list(data.columns)
Ind_features.remove('Delay Minutes')
x=data[Ind_features]
y=data['Delay Minutes']
```

```python
Ind_features=list(data.columns)
Ind_features.remove('Churned')
x=data[Ind_features]
y=data['Churned']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
from sklearn.linear_model import LinearRegression
```

```python
l_model=LinearRegression()
```

```python
l_model.fit(x_train,y_train)
predictions=l_model.predict(x_test)
error_pred=pd.DataFrame(columns={'Actual_data','Prediction_data'})
error_pred['Actual_data']=y_test
error_pred['Prediction_data']=predictions
error_pred['Error']=error_pred['Actual_data']-error_pred['Prediction_data']
```

```python
from sklearn.metrics import *
mean_squared_error(y_test,predictions,squared=False)
```

```python
mean_absolute_percentage_error(y_test,predictions)
```

```python
logisticRegr=LogisticRegression()
logisticRegr.fit(x_train,y_train)
test_pred=logisticRegr.predict(x_test)

```

```python
pred=pd.DataFrame()
pred['actual']=y_test
pred['prediction']=test_pred
pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
accuracy_score(y_test,test_pred)
```

```python
confusion_matrix(y_test,test_pred)
```

```python
print(classification_report(y_test,test_pred))
```
