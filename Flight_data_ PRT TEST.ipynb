```python
#A function that will create a new column - Price difference, that will contain the 
#price difference between the actual fare and the competitor fare for only the Gold status passengers.
def fun2():
    n=0
    df['Price difference']=''
    for i in df['Frequent Flyer Status']:
        if i == 'Gold':
            df['Price difference'].loc[n]=df['Ticket Price'].loc[n] -df['Competitor Price'].loc[n]
            n+=1
        else:
            df['Price difference'].loc[n]=''
            n+=1
```

```python
data=df.loc[df['Booking Class'] == 'Business']
```

```python
from scipy import stats  
from statsmodels.stats import weightstats as stests  
from scipy.stats import ttest_ind 
from scipy.stats import ttest_1samp  
```

```python
t_test, p_val = ttest_1samp(data['Delay Minutes'], 45)  
print("The P-value is: ", p_val,'\n')  
if p_val < 0.05:      
    print(" We can reject the null hypothesis")  
else:  
    print("We can accept the null hypothesis")  
```

```python
#Using python programming, calculate the probability of a business class passenger in the 
#gold status getting delayed more than 60 minutes.
```

```python
from scipy.stats import norm
import numpy as np
  
data=df.loc[(df['Booking Class'] == 'Business') & (df['Frequent Flyer Status'] == 'Gold')]
data1=data['Delay Minutes'].values.tolist()

mean = np.mean(data1)
std = np.std(data1)
#x=data.loc[data['Delay Minutes']>60]
#y=data['Delay Minutes'].values.tolist()
  
probability_pdf = norm.pdf(60, loc=mean, scale=std)
print(probability_pdf)

# As the pdf is 0.012, it shows probability of a business class passenger in the gold status are 
#getting delayed more than 60 minutes in chance of 1 %
```
